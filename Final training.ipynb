```python
# put all the pre-processed texts into a df so that we can do word_freq count 
def lemmtize_to_df(data_lemmatized):
    lemma_texts_df = pd.DataFrame(index = None,columns=['lemma_texts'])
    i = 0
    count = 0
    for text in data_lemmatized:
        print(i)
        str_lemma = ' '
        for word in text:
            str_lemma+=word
            str_lemma += ' '
            count+=1
        lemma_texts_df.loc[i] = str_lemma
        i+=1
    print(count)
    lemma_texts_df.head()
    return lemma_texts_df
```

```python
corpus, id2word, data_lemmatized = LDAVisual.prepare_document_ngram(df=df,X_col='text')
```

```python
lemma_txts_df = lemmtize_to_df(data_lemmatized)
```

```python
# plot freq word list for lemmatized list
freq_dict = LDAVisual.freq_words(doc=lemma_txts_df['lemma_texts'])
len(freq_dict)
```

```python
# preparing to put LDA model output into a df and csv file
doc_topic = [dict(lda_model[x]) for x in corpus]
doc_topic_df = pd.DataFrame(doc_topic)

bus_id = [bus_id for bus_id in df['business_id']]
review = [text for text in df['text']]
star = [star for star in df['stars']]
# pd.DataFrame(mixture).to_csv("topic_mixture.csv")
```

```python
# doc_topic_df['business_id'] = bus_id
doc_topic_df.insert(loc=0, column='business_id', value=bus_id)
doc_topic_df.insert(loc=1, column='review', value=review)
doc_topic_df.insert(loc=2, column='stars', value=star)
```

```python
column_names = ['business_id','review','star']
topic = "topic "
num_topic = 30
for i in range(num_topic):
    topic_i = topic + str(i)
    column_names.append(topic_i)
doc_topic_df = doc_topic_df.fillna(0)
doc_topic_df.columns = column_names
pd.DataFrame(doc_topic_df).to_csv("FINALL_doc_topic.csv")
```
