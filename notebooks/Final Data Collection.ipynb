```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
df_review = pd.read_csv('yelp_dataset/review.csv')
```

```python
all_business = df_review['business_id'].unique()
```

```python
# Calculting which business has the most reviews in this dataset
df_business_inReview = pd.DataFrame(index = None,columns=['business_id'])
df_business_inReview['business_id'] = df_review['business_id']
```

```python
# get reviews with two or less stars 
df_totalreviews = df_review[df_review['stars'] <= 2]
num_uniquebusinesses = df_totalreviews['business_id'].unique()
print(len(df_totalreviews))
print(len(num_uniquebusinesses))
```

```python
business_freq = df_business_inReview.business_id.value_counts()
```

```python
# get the business id of the top 500 resturants (by num of review)

ls_buz_id = []

for index, info in enumerate(business_freq.iteritems()):
    if index > 10:
        break
    else:
        buz_id = info[0]
        count = info[1]
        ls_buz_id.append(buz_id)
```

```python
low_ratings_df = pd.DataFrame(index = None,columns=['business_id','texts'])
low_ratings_df_tmp = pd.DataFrame(index = None,columns=['business_id','texts'])
```

```python
# get the reviews
count = 0
df_ls = []
for buz_id in ls_buz_id:
    print(count)
    low_ratings_df = df_review.loc[(df_review['business_id'] == buz_id) & (df_review['stars'] <= 2)]
    df_ls.append(low_ratings_df)
    count += 1
```

```python
low_ratings_df = pd.concat(df_ls).sort_index()
print(len(low_ratings_df), "reviews")
```

```python
# see how many words in total
word_len = 0
for text in low_ratings_df['text']:
    word_len += len(text.split())
print(word_len, "words")
```

```python
# randomly sort df 
low_ratings_df = low_ratings_df.sample(frac=1).reset_index(drop=True)
```

```python
low_ratings_df.to_csv('10buz.csv',index = False)
```
